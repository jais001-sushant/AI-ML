## Data Cleaning


```python
clean_data = data.copy()
```

```python
#Total Charges should be numeric amount. Let's convert it to numerical data type
```

```python
clean_data.TotalCharges = pd.to_numeric(clean_data.TotalCharges, errors='coerce')
clean_data.isnull().sum()
```

```python
# we can see there are 11 missing values in TotalCharges column. Let's check these records 
```

```python
clean_data.loc[clean_data ['TotalCharges'].isnull() == True]
```

**Missing Value Treatement**

```python
#Removing missing values 
clean_data.dropna(how = 'any', inplace = True)
```

```python
#Divide customers into bins based on tenure e.g. for tenure < 12 months: assign a tenure group if 1-12, for tenure between 1 to 2 Yrs, tenure group of 13-24; so on...
```

```python

print(clean_data['tenure'].max()) 
```

```python
# Group the tenure in bins of 12 months
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]

clean_data['tenure_group'] = pd.cut(clean_data.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
clean_data['tenure_group'].value_counts()
```

```python
#drop column customerID and tenure
clean_data.drop(columns= ['customerID','tenure'], axis=1, inplace=True)
clean_data.head()
```

## EDA (Exploratory Data Analysis)
Plot distibution of individual predictors by churn

### Univariate Analysis

```python
for i, predictor in enumerate(clean_data.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data=clean_data, x=predictor, hue='Churn')
```

 Convert the target variable 'Churn'  in a binary numeric variable i.e. Yes=1 ; No = 0

```python
clean_data['Churn'] = np.where(clean_data.Churn == 'Yes',1,0)
```

```python
clean_data.head()
```

Convert all the categorical variables into dummy variables/one-hot encoding

```python
clean_data_dummies = pd.get_dummies(clean_data)
clean_data_dummies.head()
```

Relationship between Monthly Charges and Total Charges

```python
sns.lmplot(data=clean_data_dummies, x='MonthlyCharges', y='TotalCharges', fit_reg=False)
```

Total Charges increase as Monthly Charges increase - as expected.

 Churn by Monthly Charges and Total Charges

```python
Mth = sns.kdeplot(clean_data_dummies.MonthlyCharges[(clean_data_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Mth = sns.kdeplot(clean_data_dummies.MonthlyCharges[(clean_data_dummies["Churn"] == 1) ],
                ax =Mth, color="Blue", shade= True)
Mth.legend(["No Churn","Churn"],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charges')
Mth.set_title('Monthly charges by churn')
```

By this we can say: Churn is high when Monthly Charges are high

```python
Tot = sns.kdeplot(clean_data_dummies.TotalCharges[(clean_data_dummies["Churn"] == 0) ],
                color="Red", shade = True)
Tot = sns.kdeplot(clean_data_dummies.TotalCharges[(clean_data_dummies["Churn"] == 1) ],
                ax =Tot, color="Blue", shade= True)
Tot.legend(["No Churn","Churn"],loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Total Charges')
Tot.set_title('Total charges by churn')
```

 Higher Churn at lower Total Charges

However if we combine the insights of 3 parameters i.e. Tenure, Monthly Charges & Total Charges then the picture is bit clear :- Higher Monthly Charge at lower tenure results into lower Total Charge. Hence, all these 3 factors viz **Higher Monthly Charge**,  **Lower tenure** and **Lower Total Charge** are linkd to **High Churn**.

 Building a corelation of all predictors with 'Churn' 

```python
plt.figure(figsize=(20,8))
clean_data_dummies.corr()['Churn'].sort_values(ascending = False).plot(kind='bar')
```

**Derived Insights:**

**HIGH** Churn seen in case of  **Month to month contracts**, **No online security**, **No Tech support**, **First year of subscription** and **Fibre Optics Internet**

**LOW** Churn is seens in case of **Long term contracts**, **Subscriptions without internet service** and **The customers engaged for 5+ years**

Factors like **Gender**, **Availability of PhoneService** and **# of multiple lines** have alomost **NO** impact on Churn

This is also evident from the **Heatmap** below

```python
# Heatmap for correlation
plt.figure(figsize=(12,12))
sns.heatmap(clean_data_dummies.corr(), cmap="Paired")
plt.title("Feature Correlation with Churn")
```
